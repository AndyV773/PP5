## Correlation Study

```python
df_corr = df.copy()
```

```python
df_corr.isna().sum()
```

```python
df_corr.info()
```

```python
df_corr['date'] = pd.to_datetime(df_corr['date'], utc=True)
```

```python
df_corr.info()
```

```python
df_corr.head(3)
```

Spearman

```python
corr_spearman = df_corr.corr(method='spearman')['target'].sort_values(key=abs, ascending=False)[1:].head(10)
corr_spearman
```

Pearson

```python
corr_pearson = df_corr.corr(method='pearson')['target'].sort_values(key=abs, ascending=False)[1:].head(10)
corr_pearson
```
